```python
X_train = train.drop(['ID','전화해지여부'], axis=1)
y_train = train['전화해지여부']
test_x = test.drop(['ID'],axis=1)
```

```python
X_train
```

```python
from supervised.automl import AutoML
```

```python
automl = AutoML(mode="Compete", eval_metric='f1', golden_features=False,ml_task='binary_classification',features_selection = False,validation_strategy={"validation_type": "kfold", "k_folds": 5} # K-Fold 횟수 설정
,total_time_limit=3600*3)
```

```python
automl.fit(X_train, y_train)
```

```python
pred = automl.predict(test_x)
```
