```python
#image = cv2.imread("경로")
#이미지를 grayscale을 해준다
#rgb에 비해 계산량 감소
image = cv2.imread('/content/KakaoTalk_20220925_210324840_05.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#pytesseract는 numpy를 인식하지 못하기 때문에 os로 파일을 불러야 한다.
filename = '{}.png'.format(os.getpid())
cv2.imwrite(filename,gray)
```

```python
text = pytesseract.image_to_string(Image.open(filename), lang = None)
os.remove(filename)
```
