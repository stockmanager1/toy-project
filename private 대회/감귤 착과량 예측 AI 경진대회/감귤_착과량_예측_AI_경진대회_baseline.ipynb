```python
import autogluon.core as ag
import pandas as pd
import torch
import torch.nn as nn
import xgboost
import lightgbm
# 모든 모듈 다시 import
from autogluon.tabular import TabularDataset, TabularPredictor
from autogluon.tabular.models import CatBoostModel, LGBModel, RFModel, XGBoostModel
```

```python
X_train = train.drop(['ID','착과량(int)'], axis=1)
y_train = train['착과량(int)']
```

```python
X_train['target'] = y_train
```

```python
X_train
```

```python
predictor = TabularPredictor(label='target', eval_metric='mean_absolute_error').fit(
    train_data=X_train, 
    num_gpus=1,  # GPU 사용 여부를 지정합니다.
    presets='best_quality', 
    time_limit=5000
)
learner_type = predictor._learner_type
```

```python
predictor.leaderboard(silent = True)
```

```python
test
```

```python
test_x = test.drop(columns=['ID'])
```

```python
model_to_use = predictor.get_model_best()
model_pred = predictor.predict(test_x, model=model_to_use)
```

```python
submit = pd.read_csv('./sample_submission.csv')
submit['착과량(int)']=np.array(model_pred)

submit.head()

submit.to_csv('./  no.csv', index=False)
```
