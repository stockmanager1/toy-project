```python
def euclidean_distance(coord1, coord2):
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    return math.sqrt((lat2 - lat1) ** 2 + (lon2 - lon1) ** 2)
```

```python
target_coord = (37.512, 126.6731)

# Í∞Å ÏÑúÏÜåÏôÄÏùò Í±∞Î¶¨ Í≥ÑÏÇ∞
result1['Í±∞Î¶¨'] = result1.apply(lambda row: euclidean_distance((row['ÏúÑÎèÑ'], row['Í≤ΩÎèÑ']), target_coord), axis=1)

# Í±∞Î¶¨Ïóê Îî∞Îùº Ï†ïÎ†¨ÌïòÍ≥† ÏÉÅÏúÑ 5Í∞ú ÏÑ†ÌÉù
closest_fire_stations = result1.nsmallest(5, 'Í±∞Î¶¨')

print("Í∞ÄÏû• Í∞ÄÍπåÏö¥ 5Í∞úÏùò ÏÑúÏÜå:")
print(closest_fire_stations[['ÏÑúÏÜåÏù¥Î¶Ñ', 'Í±∞Î¶¨']])
```
