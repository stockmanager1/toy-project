```python
def euclidean_distance(coord1, coord2):
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    return math.sqrt((lat2 - lat1) ** 2 + (lon2 - lon1) ** 2)
```

```python
target_coord = (37.512, 126.6731)

# 각 서소와의 거리 계산
result1['거리'] = result1.apply(lambda row: euclidean_distance((row['위도'], row['경도']), target_coord), axis=1)

# 거리에 따라 정렬하고 상위 5개 선택
closest_fire_stations = result1.nsmallest(5, '거리')

print("가장 가까운 5개의 서소:")
print(closest_fire_stations[['서소이름', '거리']])
```
