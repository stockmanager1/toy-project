#상관관계 분석
바로 corr을 이용후 히트맵으로 표현하려니까 좀 결측치가 많아서 깨짐. 그래서 일단 표로 표현을 했다

```python
train_corr_df=train_df.corr()
```

```python
train_df.corr()
```

그래서 일단 copy를 뜨고 결측치가 대부분을 차지하는 연결로 코드,	최고속도제한, 시작점의 위도, 통과제한 높이를 제거한 이후 다시 상관관계를 분석해보자.

```python
train_copy_df = train_df.copy()
```

```python
train_copy_df
```

```python
del_list =  ['연결로 코드','최고속도제한','시작지점의 위도','통과제한높이' ]
```

```python
for i in del_list:
  train_copy_df.drop(columns = [i])
```

```python
train_copy_df.drop(columns = '시작지점의 위도')
```

```python
train_copy_df
```

```python
train_copy_del_df=train_copy_df.drop(del_list, axis = 1)
```

```python
train_copy_del_df
```

```python
print('삭제이전:', len(train_df.columns))
```

```python
print('삭제이후:', len(train_copy_del_df.columns))
```

```python
train_copy_del_df.corr()
```

```python
#상관분석하기
#각 변수가 어떤 상관 관계가 있는지. 수치가 높을수록 관계가 있다는 것이다.
fig = plt.figure(figsize = (10,10))
sns.heatmap(train_copy_del_df.corr(), annot = True)
plt.show()
```
